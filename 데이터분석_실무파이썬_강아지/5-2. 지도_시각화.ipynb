```python
import requests

searching = 'í•©ì • ìŠ¤íƒ€ë²…ìŠ¤'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

headers = {
    'Authorization': "KakaoAK *******" # REST API
}

places = requests.get(url, headers = headers).json()['documents']
#places = url ì£¼ì†Œì— ì •ë³´ë¥¼ ìš”ì²­(headersì— API í‚¤ ê°’ì„ í•¨ê»˜ ì „ë‹¬).ìš”ì²­ í›„ ë°›ì€ ê²°ê³¼ë¥¼ json í˜•íƒœë¡œ ë¶ˆëŸ¬ì˜´[documents í•­ëª©ì˜ ê°’ì„ ì„ íƒ]
places
# ì •í™•ë„ ë†’ì€ ìˆœ
```

```python
def find_places(searching):
    # 1. ì ‘ì† url ë§Œë“¤ê¸°
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    
    # 2. headers ì…ë ¥í•˜ê¸°
    headers = {
    'Authorization': "KakaoAK *******" # REST API
    }
    
    # 3. API ìš”ì²­ ë° ì •ë³´ ë°›ê¸°
    places = requests.get(url, headers = headers).json()['documents']
    
    # 4. í•„ìš”í•œ ì •ë³´ ì„ íƒí•˜ê¸°
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]
    
    return data
```

```python
#from tqdm.notebook import tqdm # ì§„í–‰ë¥ , ì§„í–‰ì‹œê°„ í‘œì‹œ
from tqdm import tqdm
```

```python
import time
locations_inform = [ ]
for location in tqdm(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass
locations_inform
```

```python
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['ë„¤ì´ë²„ìœ„ì¹˜ëª…','ê²½ë„','ìœ„ë„','ì¸ìŠ¤íƒ€ìœ„ì¹˜ëª…']
#locations_inform_df.to_excel('5.locations.xlsx', index=False)
locations_inform_df.head()
```

```python
from folium.plugins import MarkerCluster

locations = [ ]
names = [ ]

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append([data['ìœ„ë„'], data['ê²½ë„']]) # ìœ„ë„, ê²½ë„ ìˆœìœ¼ë¡œ...
    names.append(data['ë„¤ì´ë²„ìœ„ì¹˜ëª…'])
    
Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)

marker_cluster = MarkerCluster(locations = locations, 
                                popups = names,
                                name = 'Jeju',
                                overlay = True,
                                control = True)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```
