### 카카오 로컬 API를 활용한 장소 검색

```python
import requests

searching = '합정 스타벅스'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

headers = {
    'Authorization': "KakaoAK *******" # REST API
}

places = requests.get(url, headers = headers).json()['documents']
#places = url 주소에 정보를 요청(headers에 API 키 값을 함께 전달).요청 후 받은 결과를 json 형태로 불러옴[documents 항목의 값을 선택]
places
# 정확도 높은 순
```

```python
def find_places(searching):
    # 1. 접속 url 만들기
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    
    # 2. headers 입력하기
    headers = {
    'Authorization': "KakaoAK *******" # REST API
    }
    
    # 3. API 요청 및 정보 받기
    places = requests.get(url, headers = headers).json()['documents']
    
    # 4. 필요한 정보 선택하기
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]
    
    return data
```

```python
data = find_places('제주공항')
data
```

- 인스타그램 위치명에 대한 위치 정보 검색

```python
#from tqdm.notebook import tqdm # 진행률, 진행시간 표시
from tqdm import tqdm
```

```python
import time
locations_inform = [ ]
for location in tqdm(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass
locations_inform
```

```python
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['네이버위치명','경도','위도','인스타위치명']
#locations_inform_df.to_excel('5.locations.xlsx', index=False)
locations_inform_df.head()
```

```python
location_counts_df = pd.read_excel('5.location_counts.xlsx', index_col = 0)
location_counts_df.head()
```

```python
locations_inform_df = pd.read_excel('5.locations.xlsx')
locations_inform_df.head()
```

```python
location_data = pd.merge(locations_inform_df, location_counts_df,
                         how = 'inner', left_on = '네이버위치명', right_index = True)
location_data.head()
```

```python
# 중복 데이터 점검
location_data['네이버위치명'].value_counts()
```

```python
location_data = location_data.pivot_table(
    index = ['네이버위치명','경도','위도'],
    values = 'place',
    aggfunc = 'sum')
location_data
```

```python
from folium.plugins import MarkerCluster

Mt_Hanla = [33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

#tiles = ['stamewatercolor', 'cartodbpositron','openstreetmap','stamenterrain']
folium.TileLayer('stamenterrain').add_to(map_jeju)

for i in range(len(location_data)):
    name = location_data['네이버위치명'][i] # 공식 명칭
    count = location_data['place'][i]  # 게시글 개수
    size = int(count)*2
    long = float(location_data['위도'][i])
    lat = float(location_data['경도'][i])
    folium.CircleMarker(location =[long, lat], radius = size, color = 'red', popup = name).add_to(map_jeju)

map_jeju
```

```python
map_jeju.save('jeju.html')
```

### 그룹으로 표시

```python
from folium.plugins import MarkerCluster

locations = [ ]
names = [ ]

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append([data['위도'], data['경도']]) # 위도, 경도 순으로...
    names.append(data['네이버위치명'])
    
Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)

marker_cluster = MarkerCluster(locations = locations, 
                                popups = names,
                                name = 'Jeju',
                                overlay = True,
                                control = True)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```

```python
map_jeju2.save('jeju_cluster.html')
```

```python
# 여러 개의 단어 선택/추출./저장하기
select_word_list = ['해돋이','박물관','힐링','게스트하우스','섭지코지']

for select_word in select_word_list:
    check_list = []
    for content in raw_total['content']:
        if select_word in content:
            check_list.append(True)
        else:
            check_list.append(False)
            
    select_df = raw_total[check_list]
    fpath = f'5.select_data_{select_word}.xlsx'
    select_df.to_excel(fpath)
```
