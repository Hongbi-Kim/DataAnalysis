## 전처리

```python
import pandas as pd
```

```python
# 서울시 스타벅스 매장 목록 엑셀
seoul_starbucks = pd.read_excel('6.seoul_starbucks.xlsx', header = 0)
seoul_starbucks.head()
```

```python
# 스타벅스 주소 정보에서 시군구명 추출
sgg_names = []
for address in seoul_starbucks['주소']:
    sgg = address.split(' ')[1]
    sgg_names.append(sgg)
seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()
```

## 스타벅스 분석 데이터 만들기

```python
# 시군구 목록 데이터
seoul_sgg = pd.read_excel('6.seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
# 스타벅스 매장 목록 데이터
seoul_starbucks.head()
```

```python
# 시군구별 매장수 세기
starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구명', values = '매장명', aggfunc = 'count')
starbucks_sgg_count.rename(columns = {'매장명' : '스타벅스_매장수'})
starbucks_sgg_count.head()
```

```python
# 서울시 시군구 목록 데이터에 스타벅스 매장 수 데이터를 평합
seoul_sgg = pd.merge(seoul_sgg, starbucks_sgg_count, how = 'left', on = '시군구명')
seoul_sgg.head()
```

```python
# 서울시 시군구별 인구통계 데이터
seoul_sgg_pop = pd.read_excel('6.sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop, how = 'left' , on = '시군구명')
seoul_sgg.head()
```

```python
# 시군구별 사업체 수 통계 데이터 병햡
seoul_sgg_biz = pd.read_excel('6.sgg_biz.xlsx')
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_biz, how = 'left', on = '시군구명')
seoul_sgg.head()
```
