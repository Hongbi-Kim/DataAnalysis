```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import numpy as np
```

## 멜론 크롤링

```python
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service = service)
```

```python
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
song_data = []
rank = 1
songs = soup.select('tbody > tr')
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > span > a')[0].text
    song_data.append(['Melon',rank, title, singer])
    rank += 1
song_data
```

- song_data 리스트를 이용해 데이터프레임 만들기

```python
columns = ['서비스','순위','타이틀','가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.head()
```

- 크롤링 결과를 엑셀 파일로 저장

```python
pd_data.to_excel('melon.xlsx', index=False)
```

## 지니 크롤링

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

service = Service('chromedriver.exe')
driver = webdriver.Chrome(service = service)
```

```python
url = "https://mw.genie.co.kr/chart"
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('ul.list-chart > li')
print(len(songs))
songs
```

```python
song
```

```python
song = songs[0]
title = song.select('div.info p')[0].text.strip() # strip() : 공백 문자와 줄바꿈 기호 삭제
title
```

```python
singer = song.select('div.info span')[0].text
singer
```

```python
for song in songs:
    title = song.select('div.info p')[0].text.strip()
    singer = song.select('div.info span')[0].text
    print(title, singer, sep = '  :  ')
```

```python
song_data = []
rank = 1
for song in songs:
    title = song.select('div.info p')[0].text.strip()
    singer = song.select('div.info span')[0].text
    song_data.append(['Genie', rank, title, singer])
    rank += 1
```

```python
columns = ['서비스','순위','타이틀','가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.to_excel('genie.xlsx', index=False)
```

# 멜론, 벅스, 지니 파일 통합하기

```python
excel_names = ['melon.xlsx','bugs.xlsx','genie.xlsx']
appended_data = pd.DataFrame()
for name in excel_names:
    pd_data = pd.read_excel(name) # excel 파일 내용을 읽고, 변수에 추가
    appended_data = appended_data.append(pd_data)
```
