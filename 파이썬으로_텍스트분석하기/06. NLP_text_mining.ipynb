```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('CEO3.csv', encoding = 'latin')
df[0:10]
```

```python
texts = df.text
texts
```

```python
from nltk.tokenize import RegexpTokenizer # 정규표현식 사용 정제
tokenizer = RegexpTokenizer(r'\w+')
# + : 영문자, 숫자와 공백 제외한 모든 문자 제거
```

```python
from nltk.stem import WordNetLemmatizer # 표제어
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = nltk.corpus.stopwords.words('english')
```

```python
stop_words.extend([])
```

```python
stop_words[:10]
```

```python
from nltk import wordnet
nltk.download('wordnet')
```

```python
preprocessed_texts = [ ]

for text in texts.values:
    tokenized_text = tokenizer.tokenize(text.lower()) # 소문자로 변경
    lemmatized_text = [lemmatizer.lemmatize(token) for token in tokenized_text] 
    # 토큰화된 텍스트를 표제어로 구성된 리스트 생성
    stopped_text = [token for token in lemmatized_text if token not in stop_words]
    # 불용어가 아닌 토큰이라면, 표제어 리스트에서 토큰을 추출해 리스트로 만들기
    preprocessed_texts.append(stopped_text)
```

```python
preprocessed_texts[0]
```

```python
from collections import Counter
tokens = []
for text in preprocessed_texts:
    tokens.extend(text)
counted_tokens = Counter(tokens) # 전처리된 토큰리스트를 카운팅 함
top_20 = counted_tokens.most_common(20)
top_20
```

```python
posts = df.본문
```

```python
from eunjeon import Mecab
tagger = Mecab()
```

```python
stop_words = "은 이 것 등 더 를 좀 즉 인 옹 때 만 원 이때 개 일 기 시 럭 갤 성 삼 스 폰 트 드 기 이 리 폴 사 전 마 자 플 블 가 "
```

```python
stop_words = stop_words.split(' ')
stop_words[0:10]
```

```python
nouns = [ ]
for post in posts:
    for noun in tagger.nouns(post):
        if noun not in stop_words:
            nouns.append(noun)        
```

```python
nouns
```

```python
nouns[:20]
```

```python
words = [ ]
for post in posts:
    words.extend(tagger.pos(post)) # pos : part of speech(품사)
```

```python
words[:10]
```

```python
verbs = [ ]
for word in words:
    if word[1] == 'VV':
        verbs.append(word[0] + '다')
```

```python
word[0], word[1] # word[0] : 단어, word[1] : 단어에 tagging 된 품사
```

```python
verbs[:10]
```

```python
adjective = [ ]
for word in words:
    if word[1] == 'VA':
        adjective.append(word[0] + '다')
```
