```python
import csv
import matplotlib.pyplot as plt
```

```python
f= open("seoul_book.csv")
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
            
plt.hist(result, bins=100,color='r')
plt.show()
```

### 8월의 최고 기온

```python
f= open("seoul_book.csv")
data = csv.reader(f)
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
            
plt.hist(aug, bins=100,color='r')
plt.show()
```

#### 과제
- 1월 최저기온 히스토그램으로 표현하기

```python
f = open('seoul_book.csv')
data = csv.reader(f)
next(data)
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-2] != '':
        if month == '01':
            jan.append(float(row[-2]))
            
plt.hist(jan, bins=100, color='r')
plt.show()
```

### 1월과 8월 최고기온 히스토그램

```python
f = open("seoul_book.csv")
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))
            
plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.legend()
plt.show()
```

```python
f= open("seoul_book.csv")
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

```python
f= open("seoul_book.csv")
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '06':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.boxplot(day)
plt.show()
```

```python
f= open("seoul_book.csv")
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '06':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300) # 그래프 크기 수정
plt.boxplot(day, showfliers=False) # 이상치 출력x
plt.show()
```
