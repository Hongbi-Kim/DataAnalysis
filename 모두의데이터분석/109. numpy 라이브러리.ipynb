## 1. matplotlib

```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i/10)**2)
    p3.append((i/10)**3)
plt.plot(t,t,'r--',t,p2,'bs',t,p3,'g^')
plt.show()
```

### mask
- 어떤 조건에 부합하는 데이터만 선별적으로 저장하는 기능

```python
a = np.arange(-5, 5)
print(a)
print(a<0) # 0보다 작은 값인지 확인. 배열 전체에 조건 적용
print(a[a<0]) # 다시 배열에 적용하면 조건에 부합하는 데이터만 남는다
```

```python
# mask를 변수에 넣어서 사용 가능
mask1 = abs(a) > 3 # a에 저장된 원소의 절대값이 3보다 크다. 절대값을 구해주는 absolute value 함수
print(a[mask1])  
```

```python
# 여러 마스크를 연결해 사용 가능
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1 + mask2]) # 두 조건 중 하나라도 참일 경우
print(a[mask1 * mask2]) # 두 조건 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤값 추출
y = np.random.randint(-100, 100, 1000)
mask1 = abs(x) > 50 # 절대값이 50보다 크면 거름
mask2 = abs(y) > 50

x=x[mask1+mask2]
y=y[mask1+mask2] # mask 1, 2 중 하나라도 만족하는 값 저장

#size = np.random.rand(len(x)) * 100 # 추천 코드
size = np.random.rand(1) * 100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤값 추출
y = np.random.randint(-100, 100, 1000)
mask1 = abs(x) > 50 # 절대값이 50보다 크면 거름
mask2 = abs(y) > 50

x=x[mask1+mask2]
y=y[mask1+mask2] # mask 1, 2 중 하나라도 만족하는 값 저장

size = np.random.rand(len(x)) * 100 # 추천 코드
#size = np.random.rand(1) * 100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

- 코드가 단축돼 깔끔해진다
- for 함수의 경우 반복이 줄어들어 처리 속도가 빨라진다
- 배열에 연산이나 함수를 적용하면 배열의 모든 값이 한꺼번에 계산되므로 대량의 데이터에 대한 작업이 용이해진다
