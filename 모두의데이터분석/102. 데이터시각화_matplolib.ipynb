```python
import csv
f = open("seoul_new.csv")
seoul = csv.reader(f)
next(seoul)
high = []
low = []

for row in seoul:
    if row[-1] != '' and row[-2] != '':
        if 1983 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.plot(high,'hotpink')
plt.plot(low,'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("seoul_new.csv")
seoul = csv.reader(f)
next(seoul)
high = []
low = []

for row in seoul:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if 1983 <= int(date[0]):
            if date[1] == '06' and date[2] == '22':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font', family="Malgun Gothic")
plt.rcParams['axes.unicode_minus'] = False #    
plt.title('    ')
                
plt.plot(high,'hotpink', label = 'high')
plt.plot(low,'blue', label = 'low')
plt.legend()
plt.show()
```
